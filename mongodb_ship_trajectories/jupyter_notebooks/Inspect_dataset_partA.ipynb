#### folder noaa_weather

```python
##
##weather 2019 Jan 

weather_dir = os.path.join(DATASETS, "noaa_weather", "2019", "jan")
print("weather_dir:", weather_dir)

files = sorted(os.listdir(weather_dir))
files
```

```python
#### weather 2019 Jan -> PRJ


prj_files = [f for f in os.listdir(weather_dir) if f.endswith(".prj")]
print("prj_files:", prj_files)

prj_path = os.path.join(weather_dir, prj_files[0])
with open(prj_path, "r", encoding="utf-8", errors="ignore") as f:
    print(f.read())
```

```python
#### cpg file


cpg_files = [f for f in files if f.endswith(".cpg")]
print("cpg_files:", cpg_files)

if cpg_files:
    cpg_path = os.path.join(weather_dir, cpg_files[0])
    with open(cpg_path, "r", encoding="utf-8", errors="ignore") as f:
        print(f.read())
```

```python
shp_files = [f for f in files if f.endswith(".shp")]
dbf_files = [f for f in files if f.endswith(".dbf")]
shx_files = [f for f in files if f.endswith(".shx")]

print("shp:", shp_files)
print("dbf:", dbf_files)
print("shx:", shx_files)
```

```python
for f in files:
    path = os.path.join(weather_dir, f)
    size_mb = os.path.getsize(path) / (1024*1024)
    print(f"{f:35s}  {size_mb:.2f} MB")
```

```python
import geopandas as gpd

shp_path = os.path.join(weather_dir, shp_files[0])
gdf = gpd.read_file(shp_path)

print("rows:", len(gdf))
print("columns:", gdf.columns.tolist())
print("crs:", gdf.crs)
print("geom types:", gdf.geometry.geom_type.unique())

gdf.head(5)
```

```python
weather_dir = os.path.join(DATASETS, "noaa_weather", "2019", "jan")
shp = [f for f in os.listdir(weather_dir) if f.endswith(".shp")][0]
shp_path = os.path.join(weather_dir, shp)

w = gpd.read_file(shp_path)
print("rows:", len(w))
print("columns:", list(w.columns))
print("crs:", w.crs)
print("geom:", w.geometry.geom_type.unique())
w.head(3)
```
